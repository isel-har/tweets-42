```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from preprocessor import Preprocessor


pre_proccesses =[
    {'method': None, 'vectorization':'binary'},
    {'method':'stem', 'vectorization':'binary'},
    {'method':'stem', 'vectorization':'bow'},
    {'method':'lemmatize', 'vectorization':'bow'},
    {'method':'stem', 'vectorization':'tf-idf'},
    {'method':'stem+', 'vectorization':'tf-idf'},
    {'method':'lemmatize', 'vectorization':'tf-idf', 'stopwords':True},
    {'method':'lemmatize', 'vectorization':'word2vec', 'stopwords':True},
]
# load train test split of raw tweets
data = None
try:
    data = np.load("dataset_split.npz")
except:
    pass
if data is None:
    raise SystemExit
    
X_train = data["X_train"]
X_test  = data["X_test"]

tweets = np.concatenate([X_train, X_test])

for proc in pre_proccesses:
    X = Preprocessor.process(tweets, proc)
    
    cos_sim = cosine_similarity(X)
    pairs = []

    n = len(tweets)
    for i in range(n):
        for j in range(i + 1, n):  # avoid duplicates & self-pairs
            pairs.append((i, j, cos_sim[i, j]))
    
    # # sort by similarity (descending)
    top_10 = sorted(pairs, key=lambda x: x[2], reverse=True)[:10]
    print(f"top 10 similar pair tweets using method {proc['method']}, vectorization {proc['vectorization']}")
    for i, j, score in top_10:
        print(f"[{score:.3f}]")
        print("Tweet A:", tweets[i])
        print("Tweet B:", tweets[j])
        print("-" * 50)
```
